#### 1.2 Addartist

* plt.tile：在当前图形中添加标题，可以指定标题的名称，颜色，位置和字体大小等参数
+ plt.xlabel：在当前图形中添加x轴的名称，可以指定位置、颜色和字体大小等参数
+ plt.ylabel：在当前图形中添加y轴的名称，可以指定位置、颜色和字体大小等参数
+ plt.xlim：指定当前图形x轴的范围，只能确定一个数值区间，无法使用字符串
+ plt.ylim：指定当前图形y轴的范围，只能确定一个数值区间，无法使用字符串
+ plt.xticks：指定x轴刻度的数目和取值
+ plt.yticks：指定y轴刻度的数目和取值
- plt.legend：指定当前图形的图例，可以指定图例的大小，位置和标签等等

#### 1.3、绘图

#### 1.4 图像的保存和显示

* plt.savefig：保存绘制的图形，可以指定图形的分辨率，边缘颜色等

```python
x = np.arange(0, 5, 0.2)
plt.plot(x, x)
plt.plot(x, x**2, color='g', linestyle='--', marker='*')
plt.plot(x, np.sin(x), 'rs')
plt.title('The figure of function with x')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 4.5)
plt.ylim(-5, 25)
plt.xticks([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5])
plt.yticks([-5, 0, 5, 10, 15, 20, 25])
plt.legend(['$y=x$', '$y=x^2$', '$y=sin(x)$'])
```

注：如果图中需要使用中文，则需要设置字体相关参数

```python
x = np.linspace(-2*np.pi, 4*np.pi)  # 生成x轴数据
y = np.sin(x)  # 生成y轴数据
plt.plot(x, y, label="$sin(x)$")  # 绘制sin曲线图
plt.title('$sin(x)$')
plt.xlabel('$x$轴')
plt.ylabel('$y$轴')
```

```python
# 代码 3-2
rad = np.arange(0, np.pi*2, 0.01)
# 第一幅子图
p1 = plt.figure(dpi=200)  # 确定画布大小
ax1 = p1.add_subplot(1, 2, 1)  # 创建一个两行1列的子图，并开始绘制第一幅
plt.title('lines')  # 添加标题
plt.xlabel('x')  # 添加x轴的名称
plt.ylabel('y')  # 添加y轴的名称
plt.xlim((0, 1))  # 确定x轴范围
plt.ylim((0, 1))  # 确定y轴范围
plt.xticks([0, 0.2, 0.4, 0.6, 0.8, 1])  # 规定x轴刻度
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])  # 确定y轴刻度
plt.plot(rad, rad**2)  # 添加y=x^2曲线
plt.plot(rad, rad**4)  # 添加y=x^4曲线
plt.legend(['y=x^2', 'y=x^4'])

# 第二幅子图
ax2 = p1.add_subplot(1, 2, 2)  # 创开始绘制第2幅
plt.title('sin/cos')  # 添加标题
plt.xlabel('rad')  # 添加x轴的名称
plt.ylabel('value')  # 添加y轴的名称
plt.xlim((0, np.pi*2))  # 确定x轴范围
plt.ylim((-1, 1))  # 确定y轴范围
plt.xticks([0, np.pi/2, np.pi, np.pi*1.5, np.pi*2])  # 规定x轴刻度
plt.yticks([-1, -0.5, 0, 0.5, 1])  # 确定y轴刻度
plt.plot(rad, np.sin(rad))  # 添加sin曲线
plt.plot(rad, np.cos(rad))  # 添加cos曲线
plt.legend(['sin', 'cos'])
plt.show()
```

#### 画散点图：scatter()

* plt.scatter(x, y, s=None, c=None, marker=None, alpha=None, **kwargs)
+ s:控制每个散点的大小，由数值给出
+ c：控制每个散点的颜色，由数值给出，也可以由颜色标记给出
- marker：散点的图例。常用图例可以百度查看

```python
x = np.arange(0, 10, 0.5)
y = x**2+1
p3 = plt.figure(figsize=(10, 5))
p3.add_subplot(1, 2, 1)
# 从直线y=2x+1上抽取21个散点，绘制其散点图
plt.scatter(x, y, s=5, c='r', marker='o')
plt.title('直线$y=e^x+1$上点的散点图1')
plt.xlabel('x轴')
plt.ylabel('y轴')

# 如果每个散点的大小，颜色想根据对应y的取值自行控制，可以设置s=y,c=y
p3.add_subplot(1, 2, 2)
plt.scatter(x, y, s=y, c=y, marker='o')
plt.title('直线$y=e^x+1$上点的散点图2')
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.show()
```

```python
#分类数据的例子
#从标准正态分布中随机抽取400个随机数，排成2列，每一行对应一个散点
#前100行每个分量加2，对应的类标签y值为1；后100行每个分量-2，对应y值为2
#从而得到属于1和2两类的散点
import random
x = np.random.randn(200,2)
x[0:100,:] = x[0:100,:]+1
x[100:200,:] = x[100:200,:]-1
yy = []
for i in np.arange(200):
        if i<100:
            yy.append(1)
        else:
            yy.append(2)
y = np.array(yy)
```

#### 画条形图（柱状图）

* plt.bar(x，height，width = 0.8，color,bottom = None，hold = None，data = None，** kwargs
+ x：x轴坐标轴
+ height：接受实数或者实数向量，给定每个柱子的高度
+ width：接受实数或者实数向量，给定每个柱子的宽度
+ bottom：y轴的起始值
+ color：柱体颜色
- label：标签


例：教材例5-3

```python
# 导入绘图模块
import matplotlib.pyplot as plt
import numpy as np
# 构建数据
Y2016 = [15600, 12700, 11300, 4270, 3620]
Y2017 = [17400, 14800, 12000, 5200, 4020]
labels = ['Beijing', 'Shanghai', 'HongKong', 'Shenzhen', 'Guangzhou']
bar_width = 0.5

plt.figure(figsize=(8, 5))
# 设置条形颜色、透明度、宽度
plt.bar(np.arange(5), Y2016, label='2016', color='yellow',
        alpha=0.8, width=bar_width)
plt.bar(np.arange(5) + bar_width, Y2017, label='2017',
        color='black', alpha=0.8, width=bar_width)
# 添加轴标签
plt.xlabel('Top5 City', fontproperties='Times New Roman',
           fontsize=16)  # 注意这里字体字号的设置
plt.ylabel('Family Amount', fontproperties='Times New Roman', fontsize=16)
plt.xticks(np.arange(5)+0.2, labels)
plt.ylim([2500, 20000])
plt.yticks(fontproperties='Times New Roman', size=14)
plt.xticks(fontproperties='Times New Roman', size=14)
# 添加标题
plt.title('Millions Family Amount Top5 City Distribution',
          fontproperties='Times New Roman',
          fontsize=16)
# 为每个条形图添加数值标签
for x2016, y2016 in enumerate(Y2016):
    plt.text(x2016-0.25, y2016+100, '%s' % y2016)
for x2017, y2017 in enumerate(Y2017):
    plt.text(x2017+bar_width/2, y2017+100, '%s' % y2017)
# 显示图例
plt.legend()
# plt.savefig('matplot03.png', dpi=300, bbox_inches='tight')
```

#### 绘制饼图pie()

plt.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=None, radius=None, … )

教材示例5-4

```python
# 导入绘图模块
plt.figure(figsize=(9, 6))
# 构造数据,调整数据顺序，以免低占比的饼块相邻，无法分辨
edu = [0.2515, 0.1057, 0.2724, 0.2336, 0.1368]
labels = ['HSchool', 'Others', 'College', 'Bachelor', 'Master']
explode = [0, 0, 0.1, 0, 0]  # 用于分离突出显示大专学历人群
colors = ['#6f6f6f', '#9f9f9f', '#6f6f6f', '#9f9f9f', '#cfcfcf']  # 自定义颜色
# 将横、纵坐标轴标准化处理，保证饼图是一个正圆，否则为椭圆
plt.axes(aspect='equal')
# 控制x轴和y轴的范围
plt.xlim(0, 4)
plt.ylim(0, 4)
# 绘制饼图
plt.pie(x=edu,  # 绘图数据
        explode=explode,  # 突出显示大专人群
        labels=labels,  # 添加教育水平标签
        colors=colors,  # 设置饼图的自定义填充色
        autopct='%.1f%%',  # 设置百分比的格式，这里保留一位小数
        pctdistance=0.5,  # 设置百分比标签与圆心的距离
        labeldistance=1.0,  # 设置教育水平标签与圆心的距离
        startangle=180,  # 设置饼图的初始角度
        radius=1.5,  # 设置饼图的半径
        counterclock=False,  # 是否逆时针，这里设置为顺时针方向
        wedgeprops={'linewidth': 1.5, 'edgecolor': 'green'},  # 设置饼图内外边界的属性值
        textprops={'fontsize': 15, 'color': 'k'},  # 设置文本标签的属性值
        center=(2, 2),  # 设置饼图的原点
        frame=1)  # 是否显示饼图的图框，这里设置显示

# 删除x轴和y轴的刻度
plt.xticks(())
plt.yticks(())
# 添加图标题
plt.title('XX信用违约客户分析')
plt.legend(loc="upper right", fontsize=8)
# plt.savefig('matplot04.png', dpi=300, bbox_inches='tight')
plt.show()
```
